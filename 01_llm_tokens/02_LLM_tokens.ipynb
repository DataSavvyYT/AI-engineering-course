```python
# --- 2. Import Libraries ---
from transformers import AutoTokenizer
```

```python
# --- 3. Initialize an LLM Tokenizer (BERT's WordPiece) ---
# We use the tokenizer from a popular pre-trained model (BERT-base-uncased)
# The tokenizer is the crucial component for converting text to tokens (IDs).
MODEL_NAME = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
print(f"--- Using Tokenizer: {MODEL_NAME} ---")
```

```python
# --- 4. Define Input Text Examples ---
# Example 1: Standard words
text_1 = "Tokenization is fundamental for large language models."
```

```python
# Example 2: Rare word, hyphenation, and specialized characters
# 'Subword' is a common word, but 'unpredictable' often gets split.
text_2 = "Unpredictable supercalifragilisticexpialidocious $1M"
```

```python
# --- 5. Tokenization and Encoding ---
def tokenize_and_display(input_text):
    # 1. Tokenize (convert text to the subword units)
    tokens_decoded = tokenizer.tokenize(input_text)

    print(f"\nInput Text: '{input_text}'")
    print("-" * 50)
    print("A. Subword Tokens (Decoded):")
    print(tokens_decoded)
    print(f"Total Tokens: {len(tokens_decoded)}")
```

```python
# Run the demo for the examples
tokenize_and_display(text_1)
tokenize_and_display(text_2)
```

```python
# --- 6. The Out-of-Vocabulary (OOV) Solution ---
# The word 'expialidocious' doesn't exist in BERT's vocabulary,
# so the tokenizer breaks it down into known smaller subwords.
# This prevents an OOV error.
print("\n" + "="*50)
print("## LLM Tokenization Strategy (OOV Solution)")
print("Why 'subword' is used: To handle unseen or rare words.")
unseen_word = "Supercalifragilisticexpialidocious"
unseen_tokens = tokenizer.tokenize(unseen_word)
print(f"Unseen Word: '{unseen_word}'")
print(f"Token Breakdown: {unseen_tokens}")
```
