```python
import os
import google.generativeai as genai
from google.colab import userdata
```

```python
# 1. SETUP API KEY
# You can use the "Secrets" key in Colab (left sidebar) or paste it below.
try:
    GOOGLE_API_KEY = GEMINI_API_KEY
except:
    import getpass
    GOOGLE_API_KEY = getpass.getpass("Enter your Google API Key: ")

genai.configure(api_key=GOOGLE_API_KEY)
```

```python
# 2. THE TOOL
# A simple Python function.
# IMPORTANT: Type hints (location: str) and docstrings are REQUIRED.
# The AI reads the docstring to understand what the function does.
def get_current_weather(location: str):
    """
    Returns the current weather for a specific city.

    Args:
        location: The city and state, e.g. San Francisco, CA
    """
    print(f"\n[System] ðŸ¤– AI is calling the function for: {location}...")

    # Mock data - in a real app, you'd call a real weather API here
    location = location.lower()
    if "london" in location:
        return "The weather is rainy and 15Â°C."
    elif "tokyo" in location:
        return "The weather is clear and 24Â°C."
    elif "paris" in location:
        return "The weather is sunny and 22Â°C."
    else:
        return "Weather data not available for this location."
```

```python
# 3. INITIALIZE THE AGENT
# We simply pass the function itself into the 'tools' list.
# 'gemini-1.5-flash' is fast and great for simple tasks.
model = genai.GenerativeModel(
    model_name='gemma-3-12b-it',
    tools=[get_current_weather]
)
```

```python
# 4. RUN THE CHAT
# enable_automatic_function_calling=True makes the loop invisible.
chat = model.start_chat(enable_automatic_function_calling=True)
```

```python
# Ask a question
user_query = "What is the weather like in Tokyo right now?"
print(f"User: {user_query}")

response = chat.send_message(user_query)

print(f"Agent: {response.text}")
```

```python
# 2. LIST MODELS
print("--- Available Chat Models ---\n")
for m in genai.list_models():
    # We only want models that can generate content (Chat/Text), not embedding models
    if 'generateContent' in m.supported_generation_methods:
        print(f"Name: {m.name}")
        print(f"Display Name: {m.display_name}")
        print(f"Description: {m.description}")
        print("-" * 40)
```
