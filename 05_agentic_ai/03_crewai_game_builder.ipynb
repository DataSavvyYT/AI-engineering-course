```python
inputs = {
        'game' :  examples['example3_snake']
        }
```

```python
engineering_lead = Agent(
    role = 'Senior Software Engineer',
    goal = '''Create software as needed''',
    backstory = '''You are a Senior Software Engineer at a leading tech think tank.
    Your expertise in programming in python. and do your best to produce perfect code''',
    llm = my_llm,
    verbose=True,
)
```

```python
backend_engineer = Agent(
    role = 'Software Quality Control Engineer',
    goal = '''Create Perfect code, by analyzing the code that is given for errors''',
    backstory = '''You are a software engineer that specializes in checking code
    for errors. You have an eye for detail and a knack for finding
    hidden bugs.
    You check for missing imports, variable declarations, mismatched
    brackets and syntax errors.
    You also check for security vulnerabilities, and logic errors''',
    llm = my_llm,
    verbose=True,
)
```

```python
frontend_engineer = Agent(
    role = 'Chief Software Quality Control Engineer',
    goal = '''Ensure that the code does the job that it is supposed to do''',
    backstory = '''You feel that programmers always do only half the job, so you are
    super dedicate to make high quality code.''',
    llm = my_llm,
    verbose=True,
)
```

```python
test_engineer = Agent(
    role = 'An engineer with python coding skills who can write unit tests for the given backend module {module_name}',
    goal = '''Write unit tests for the given backend module {module_name} and create a test_{module_name} in the same directory as the backend module.''',
    backstory = '''You're a seasoned QA engineer and software developer who writes great unit tests for python code.''',
    llm = my_llm,
    verbose=True,
)
```

```python

# Create tasks for your agents
design_task = Task(
  description="""You will create a game using python, these are the instructions:

    Instructions
    ------------
    {game}""",
  agent=engineering_lead,
  verbose=True,
  #output_file = "output/design.md",
  expected_output="""Your Final answer must be the full python code, only the python code and nothing else."""

)
```

```python
code_task = Task(
  description="""You will create a game using python, these are the instructions:

    Instructions
    ------------
    {game}

    Using the code you got, check for errors. Check for logic errors,
    syntax errors, missing imports, variable declarations, mismatched brackets,
    and security vulnerabilities.""",
  agent=backend_engineer,
  expected_output="Your Final answer must be the full python code, only the python code and nothing else.",
  context=[design_task],
  #output_file= "output/{module_name}"
)
```

```python
frontend_task = Task(
  description="""You are helping create a game using python, these are the instructions:

    Instructions
    ------------
    {game}

    You will look over the code to insure that it is complete and
    does the job that it is supposed to do.""",
  agent=frontend_engineer,
  expected_output="""Your Final answer must be the full python code, only the python code and nothing else.""",
  context=[code_task],
  #output_file= "output/app.py"
)
```

```python
test_task = Task(
  description="""Write unit tests for the given backend module {module_name} and create a test_{module_name} in the same directory as the backend module.""",
  agent=test_engineer,
  expected_output="""A test_{module_name} module that tests the given backend module.
    IMPORTANT: Output ONLY the raw Python code without any markdown formatting, code block delimiters, or backticks.
    The output should be valid Python code that can be directly saved to a file and executed.""",
  context=[code_task],
  output_file= "output/test_{module_name}"
)
```

```python

# Instantiate your crew with a sequential process
crew = Crew(
  agents=[engineering_lead,backend_engineer,frontend_engineer],
  tasks=[design_task,code_task,frontend_task],
  verbose=False, # You can set it to 1 or 2 to different logging levels
)
```
