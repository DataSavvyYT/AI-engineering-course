```python
from google.colab import userdata
import os
from agents import Agent, Runner, RunConfig, OpenAIProvider, WebSearchTool

# Ensure the API key is available
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
```

```python
from pydantic import BaseModel
```

```python
class VideoScript(BaseModel):
    outline: str
    script: str
```

```python
video_script_creator = Agent(
    name="Video script creator",
    model="gpt-4o-mini",
    handoff_description="Used for generating a youtube video script based on an topic outline.",
    instructions=(
        "Given a topic and an outline, your job is to generate script for creating youtube video for each section of the outline."
        "Format the script is in Markdown using a mix of text for explanation and code snippets for examples if it is coding topic."
        "make sure video script is informative, interesting and engaging"
    ),
    output_type=VideoScript
)
```

```python
outline_builder = Agent(
    name="Outline Builder",
    model="gpt-4o-mini",
    instructions=(
        "Given a particular topic, your job is to help come up with a outline for a youtube video. You will do that by crafting an outline."
        "After making the outline, hand it to the video script creator agent."
    ),
    handoffs=[video_script_creator]
)
```

```python
script_response = await Runner.run(outline_builder, "Model context protocol in Agentic AI")
print(script_response.final_output)
```
